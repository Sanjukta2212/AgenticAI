```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

# 1. Initialize the Model
model = ChatOpenAI(model="gpt-4o")

# 2. Initialize the Parser
# This parser will look for JSON code blocks in the AI's response
parser = JsonOutputParser()

# 3. Create a Prompt with Format Instructions
# The parser provides a helper to tell the AI exactly how to format the JSON
prompt = PromptTemplate(
    template="Answer the user query.\n{format_instructions}\n{query}",
    input_variables=["query"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

# 4. Chain them together
chain = prompt | model | parser

# 5. Run it
result = chain.invoke({"query": "Tell me a joke about a computer."})

# The 'result' is now a real Python Dictionary, not just a string!
# print(result["setup"])
# print(result["punchline"])
```

## Exercise: Create a simple assistant using any LLM and ChatPromptTemplate.

The assistant should take a product-related query and return structured output containing:

- product_name  
- product_details  
- price (tentative price in USD, must be an integer)

The output must follow a Pydantic schema.

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
```

### Step 1: Define Pydantic Schema

```python
class Product(BaseModel):
    product_name: str = Field(description="Name of the Product")
    product_details: str = Field(description="Short description of the product")
    price: int = Field(description="Price USD (integer only)")
```

### Step 2: Create Prompt Template

```python
template = ChatPromptTemplate([
    ("system", "You are a helpful assistant. Extract product info in structured format."),
    ("human", "{user_input}")
])
```

### Step 3: Fill the template

```python
prompt_value = template.invoke(
    {
        "user_input": "Tell me about iPhone 15"
    }
)
```

### Step 4: Load Model

```python
model = ChatOpenAI()
```

### Step 5: Attach structured output

```python
structured_model = model.with_structured_output(Product)
```

### Step 6: Call model with prompt

```python
response = structured_model.invoke(prompt_value)
```

```python
print(response)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field

class Product(BaseModel):
    product_name:str = Field(description="Name of the product")
    product_details:str = Field(description="description of the product")
    price:int = Field(description="prices in usd")

template = ChatPromptTemplate([
    ("system", "You are a doctor agent"),
    ("human", "{user_input}")
])

prompt_value=template.invoke({
    "user_input":"Tell mme about the chances of living of my Grandma who is 86 years old and she is in COMA"
})

model = ChatOpenAI()

structured_model=model.with_structured_output(Product)
```

```python
response = structured_model.invoke(prompt_value)
```

```python
print(response)
```
