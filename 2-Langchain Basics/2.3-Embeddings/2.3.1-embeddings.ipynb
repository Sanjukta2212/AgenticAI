### Embeddings

Convert Text into Vectors

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-large",
    # With the `text-embedding-3` class
    # of models, you can specify the size
    # of the embeddings you want returned.
    # dimensions=1024
)
```

```python
embeddings
```

```python
text="This is a tutoriall on Open AI Embeddings"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

Irrespective of how long the sentence is, Open AI embedding will convert it into 3072 dimensions.

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader('/Users/sanjuktabaruah/Desktop/coursera/Krishnaink/agenticAI/2-Langchain Basics/2.1-DataIngestion/speech.txt')
docs=loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
embeddings.embed_query(final_documents[0].page_content)
```
